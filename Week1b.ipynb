```python
import numpy as np
import matplotlib.pyplot as pp

def lucas(n, order=2):
    if n == 0: return [order]
    if n == 1: return [1,order]
    y = lucas(n-1, order)
    return [y[0]+y[1]]+y

lucas(100)[0]
```

```python
def mandel(c, z0):
    z = z0
    for i in range(100):
        z = z*z+c
    return (np.isnan(z) | (abs(z) > 2))*1

px = 1000
py = 1000

x = np.linspace(-1,1,px)
y = np.linspace(-1,1,py)
xx,yy = np.meshgrid(x,y)
z = xx + 1j*yy
out = mandel(z,z*0)
pp.figure(figsize=(12,12))
pp.imshow(out)
```
