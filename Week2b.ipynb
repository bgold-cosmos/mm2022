```python
import numpy as np
import matplotlib.pyplot as pp
import pandas as pd
```

```python
data = pd.read_csv('GlobalLandTemperaturesByMajorCity.csv')
```

```python
data.columns
```

```python
years = pd.DatetimeIndex(data.dt).year
months = pd.DatetimeIndex(data.dt).month
days = pd.DatetimeIndex(data.dt).day
```

```python
ii = (data.City == "Chicago") & (months == 12)
```

```python
x = years[ii]
y = data.AverageTemperature[ii]
pp.plot(x,y)
```

```python
dy = data.AverageTemperatureUncertainty[ii]
pp.figure(figsize=(16,9))
pp.errorbar(x, y, dy, fmt='k.')
pp.xlabel('year')
pp.ylabel('average temperature')
pp.title('December 1$^{st}$ temperature in Chicago')
```

```python
jj = (data.City == 'Chicago')
fig = pp.figure(figsize=(20,15))
ax = fig.add_subplot(projection='3d')
thing = ax.scatter(years[jj], months[jj], data.AverageTemperature[jj], c=data.AverageTemperatureUncertainty[jj])
fig.colorbar(thing)
```
