```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
x,y = np.meshgrid(np.linspace(-2,2,30), np.linspace(-2,2,30))

r1 = np.sqrt((x-1)**2 + y**2)
th1 = np.arctan2(y,(x-1))
u1 = np.cos(th1) / r1**2
v1 = np.sin(th1) / r1**2

r2 = np.sqrt((x+1)**2 + y**2)
th2 = np.arctan2(y,(x+1))
u2 = -np.cos(th2) / r2**2
v2 = -np.sin(th2) / r2**2

pp.figure(figsize=(10,10))
pp.quiver(x,y,u1+u2,v1+v2, pivot='mid', scale=100)
```
