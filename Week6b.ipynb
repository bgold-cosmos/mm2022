```python
import numpy as np
import matplotlib.pyplot as pp
import scipy.io.wavfile as wav
from scipy.fft import fft, ifft, fftfreq
```

```python
sr = 44100
t = np.arange(0,2,1/sr)
x = np.sin(2*np.pi* 120 * t)       # 120 Hz sine wave
y = np.mod(120*t, 1)*2-1           # 120 Hz sawtooth
z = np.round(np.mod(120*t,1))*2-1  # 120 Hz square
pp.figure(figsize=(16,9))
pp.plot(t,x,'k-', t,y,'b-', t,z,'g-')
pp.legend(["sine", "sawtooth", "square"])
pp.xlim([0,0.01])
```

```python
wav.write("sine.wav",sr,x)
wav.write("saw.wav",sr,y)
wav.write("square.wav",sr,z)
```

```python
h = np.real(ifft( fft(y) * fft(z) ))
pp.figure(figsize=(16,9))
pp.plot(t,h)
pp.xlim([0,0.02])
```
