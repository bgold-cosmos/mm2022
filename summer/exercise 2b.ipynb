```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
np.linspace(-2,2,20)
```

```python
x,y = np.meshgrid(np.linspace(-2,2,20), np.linspace(-2,2,20))
```

```python
r = np.sqrt(x**2 + y**2)
```

```python
th = np.arctan2(y,x)
```

```python
u = np.cos(th) / r**2
v = np.sin(th) / r**2
```

```python
pp.figure(figsize=(10,10))
pp.quiver(x, y, u, v, scale=100)
```
