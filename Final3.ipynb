```python
import numpy as np
import matplotlib.pyplot as pp
from scipy.integrate import solve_ivp
```

```python
from numpy import sin,cos

def derivs(t,y):
    th1 = y[0]
    th2 = y[1]
    w1 = y[2]
    w2 = y[3]
    sind = sin(th1-th2)
    cosd = cos(th1-th2)
    dw1 = sin(th2)*cosd - sind*(w1**2 * cosd + w2**2) - 2*sin(th1)
    dw1 /= 1 + sind**2
    dw2 = w1**2 * sind - sin(th2) - dw1 * cosd
    return np.array([w1,w2,dw1,dw2])
```

```python
s = solve_ivp(derivs, [0,300], [np.pi/2, np.pi/2, 0, 0], t_eval=np.arange(0,300,0.1))
```

```python
pp.plot(s.t, s.y[0], 'k-')
```
