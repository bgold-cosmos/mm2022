```python
import numpy as np
import matplotlib.pyplot as pp
from scipy.linalg import eig
```

```python
mat = np.array([[2, -1],[-1,2]])
print(mat)
e,v = eig(mat)
print(e)
pp.plot(v[:,0], 'ks', v[:,1], 'r^')
```

```python
mat = np.array([[2, -1, 0], [-1, 2, -1], [0, -1, 2]])
e, v = eig(mat)
pp.plot(np.real(e), 'k.')
pp.figure()
pp.subplot(311)
pp.plot(v[:,0], 'ks')
pp.subplot(312)
pp.plot(v[:,1], 'ks')
pp.subplot(313)
pp.plot(v[:,2], 'ks')
```

```python
mat = np.diag(2*np.ones(10))
mat += np.diag(-1*np.ones(9), 1)
mat += np.diag(-1*np.ones(9),-1)
print(mat)
```

```python
e,v = eig(mat)
sort = np.argsort(e)
pp.plot(np.real(np.sqrt(e[sort]/e[sort[0]])), 'k.')
pp.figure(figsize=(16,9))
for i in range(5):
    pp.subplot(5,1,i+1)
    pp.plot(v[:,sort[i]], 'ks-')
```
