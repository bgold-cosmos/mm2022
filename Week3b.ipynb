```python
x,y = np.meshgrid(np.linspace(-1,1,20), np.linspace(-1,1,20))

r = np.sqrt(x**2 + y**2)
th = np.arctan2(y,x)

u = np.cos(th) 
v = np.sin(th)

pp.figure(figsize=(10,10))
pp.quiver(x,y,u,v, scale=30, pivot='mid')
```

```python
import numpy as np
import matplotlib.pyplot as pp
from timeit import default_timer as timer

def timeAdd(n, reps):
    t = np.zeros(reps)
    for i in range(reps):
        a = np.random.rand(n,n)
        b = np.random.rand(n,n)
        start = timer()
        c = a + b
        end = timer()
        t[i] = end-start
    return(np.mean(t))

m = np.logspace(2,4, 20)
x = np.zeros(20)
y = np.zeros(20)

for i,s in enumerate(m):
    print(i,s)
    x[i] = int(s)
    y[i] = timeAdd(int(s), 100)

#pp.plot(x,y)
```
