```python
import numpy as np
import matplotlib.pyplot as pp
from scipy.linalg import eig
```

```python
mat = np.array([[2, -1], [-1,2]])
print(mat)
e,v = eig(mat)
print(np.real(e))
```

```python
mat = np.diag(np.ones(10)*2) + np.diag(np.ones(9)*-1, 1) + np.diag(np.ones(9)*-1, -1)
```

```python
mat
```

```python
e,v = eig(mat)
pp.plot(np.sqrt(np.real(e)), 'ks')
```

```python
e,v = eig(mat)
sort = np.argsort(e)
se = e[sort]
sv = v[:,sort]
pp.plot(np.sqrt(np.real(se)), 'ks')
```

```python
pp.plot(sv[:,0], 'ks-')
```

```python
pp.plot(sv[:,1], 'ks-')
```

```python
pp.plot(sv[:,2], 'ks-')
```

```python
pp.figure(figsize=(16,9))
for i in range(5):
    pp.subplot(5, 1, i+1)
    pp.plot(sv[:,i], 'ks-')
```
