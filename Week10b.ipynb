```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
def myfunc(x):
    return x**4 + 2*x**3 - 4*x**2 - 8*x
```

```python
x = np.linspace(-4,3,100)
pp.figure(figsize=(16,9))
pp.plot(x, myfunc(x))
```

```python
x = np.random.rand(10000)*7-4
y = myfunc(x)
x[y == np.min(y)]
```

```python
howmanytimes = 1000
xmin = np.zeros(howmanytimes)
for i in range(howmanytimes):
    x = np.random.rand(100000)*7-4
    y = myfunc(x)
    xmin[i] = x[y == np.min(y)]
```

```python
np.mean(xmin), np.std(xmin)
```
